```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats

df = pd.read_csv('/content/drive/MyDrive/FIAP/Machine Learning/atletas.csv')
del df['_tempo']
df
```

```python
df1 = df.copy()
df1['dummy_gender'] = df.apply( lambda row: 1 if row['Sexo']=='homem' else 0 , axis=1)
filtered_df = df1[df1['Altura'].notnull() & df1['Peso'].notnull() & df1['Ano de nascimento'].notnull()]
filtered_df
```

```python
x = filtered_df[['dummy_gender','Ano de nascimento','Peso','Altura']].to_numpy()
y = filtered_df['tempo'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y).round(2)
print(score)
```

```python
filtered_df = df1[ (df1['Peso'].notnull()) & (df1['Altura'].notnull())& (df1['Ano de nascimento'].notnull())]
x = filtered_df[['dummy_gender','Altura','Peso']].to_numpy()
y = filtered_df['Ano de nascimento'].to_numpy()
model_ano = LinearRegression()
model_ano.fit(x, y)
model_ano.score(x, y).round(2)
```

```python
def completar_ano(row):
  if not pd.isna(row['Peso']) and not pd.isna(row['Altura']) and pd.isna(row['Ano de nascimento']):
    campos = row[['dummy_gender','Altura','Peso']]
    return round(model_ano.predict([campos])[0])
  return row['Ano de nascimento']
df1['novo_ano'] = df1.apply(completar_ano, axis=1)
df1
```

```python
filtered_df = df1[(df1['Peso'].notnull()) & (df1['Altura'].notnull())& (df1['Ano de nascimento'].notnull())]
x = filtered_df[['dummy_gender','Ano de nascimento']].to_numpy()
y = filtered_df['Peso'].to_numpy()
model_peso = LinearRegression()
model_peso.fit(x, y)
model_peso.score(x, y).round(2)
```

```python
filtered_df = df1[(df1['Peso'].notnull()) & (df1['Altura'].notnull())& (df1['Ano de nascimento'].notnull())]
x = filtered_df[['dummy_gender','Ano de nascimento']].to_numpy()
y = filtered_df['Altura'].to_numpy()
model_altura = LinearRegression()
model_altura.fit(x, y)
model_altura.score(x, y).round(2)
```

```python
def completar_peso(row):
  if pd.isna(row['Peso']) and not pd.isna(row['Ano de nascimento']):
    campos = row[['dummy_gender','Ano de nascimento']]
    return round(model_peso.predict([campos])[0])
  return row['Peso']

df1['novo_peso'] = df1.apply(completar_peso, axis=1)
df1
```

```python
def completar_altura(row):
  if pd.isna(row['Altura']) and not pd.isna(row['Ano de nascimento']):
    campos = row[['dummy_gender','Ano de nascimento']]
    return round(model_altura.predict([campos])[0])
  return row['Altura']

df1['nova_altura'] = df1.apply(completar_altura, axis=1)
df1
```

```python
filtered_df = df1[(df1['tempo'].notnull()) & (df1['novo_ano'].notnull()) & (df1['novo_peso'].notnull()) & (df1['nova_altura'].notnull()) ]
x = filtered_df[['dummy_gender','novo_ano','novo_peso','nova_altura']].to_numpy()
y = filtered_df['tempo'].to_numpy()
model = LinearRegression()
model.fit(x, y)
model.score(x, y).round(2)
```
