```python
df_sal = pd.read_csv('/home/Salary_Data.csv')
print(df_sal.shape)
df_sal.head()
```

```python
df_sal1 = df_sal.copy()
Y = 'Salary'
X = ['YearsExperience']

model = LinearRegression()
model.fit(df_sal1[X], df_sal1[Y])
print(model.score(df_sal1[X], df_sal1[Y]))

plt.plot(df_sal1[X], df_sal1[Y], color='red')
plt.plot(df_sal1[X], model.predict(df_sal1[X]), color='blue')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

# print(df_sal1[X].to_numpy())

poly_f = PolynomialFeatures(degree=12)
X_poly = poly_f.fit_transform(df_sal1[X].to_numpy())

model.fit(X_poly, df_sal1[Y]) # observar o X tratado: será o desenho de uma curva
print(model.score(X_poly, df_sal1[Y]))

plt.plot(df_sal1[X], df_sal1[Y], color='red')
plt.plot(df_sal1[X], model.predict(X_poly), color='blue')
plt.show()
```
